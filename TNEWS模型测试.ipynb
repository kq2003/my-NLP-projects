```python
from torch.utils.data import Dataset, DataLoader

class NSPDataset(Dataset):
  def __init__(self, encoded_data):
      self.input_ids = encoded_data['input_ids']
      self.token_type_ids = encoded_data['token_type_ids']
      self.attention_mask = encoded_data['attention_mask']
      self.next_sentence_label = encoded_data['label']

  def __len__(self):
      return len(self.input_ids)

  def __getitem__(self, idx):
      return {'input_ids': self.input_ids[idx],
              'token_type_ids': self.token_type_ids[idx],
                'attention_mask': self.attention_mask[idx],
                'label': self.next_sentence_label[idx]}


train_dataset = NSPDataset(encoded_train)
dev_dataset = NSPDataset(encoded_dev)

train_dataloader = DataLoader(train_dataset, batch_size=16, shuffle=True)
dev_dataloader = DataLoader(dev_dataset, batch_size=16, shuffle=False)
```

```python
import evaluate

metric = evaluate.load('accuracy')
```

```python
def compute_metrics(eval_pred):
  predictions, labels = eval_pred
  return metric.compute(predictions=np.argmax(predictions, axis = 1),
                        references = labels)
```

```python
args = TrainingArguments(
    output_dir='/content/Drive/My Drive',
    evaluation_strategy='epoch',
    learning_rate=5e-5,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    num_train_epochs=3,
    gradient_accumulation_steps=1, #Reduced gradient accumulation
    warmup_steps=500, #Added warmup
    weight_decay=0.018, #Added weight decay
    logging_steps = 460,
    save_strategy='epoch',
)
```
